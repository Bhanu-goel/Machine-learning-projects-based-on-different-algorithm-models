CONVERT STRING INTO INTEGERS

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df["v1"]=label_encoder.fit_transform(df["v1"])
```

```python
df2=pd.value_counts(df["v1"])
df2
```

```python
df2=pd.value_counts(df["v1"])
df2.plot(kind="bar",color= ["green", "red"])
plt.title("Classification Between Ham And Spam")
plt.show()
```

```python
x=df.iloc[:,1]
y=df.iloc[:,0]
```

SPLIT THE DATA INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.shape,x_test.shape
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
x_train_new= cv.fit_transform(x_train)
```

```python
x_train_new
```

```python
y_train=y_train.astype(int)
```

```python
x_train_new=x_train_new.astype(int)
```

```python
y_test=y_test.astype(int)
```

```python
x_train_new.dtype
```

TRAIN THE DATA ON LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
```

```python
LR.fit(x_train_new,y_train)
```

```python
LR_pred=LR.predict(cv.transform(x_test))
```

ACCURACY OF DATASET

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
cm=confusion_matrix(LR_pred,y_test)
cm
```

```python
f1_score(LR_pred,y_test)
```
