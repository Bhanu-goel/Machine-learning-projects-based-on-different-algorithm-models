IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

READ THE DATA

```python
df=pd.read_csv("/kaggle/input/advertisingcsv/Advertising.csv")
```

```python
df.head()
```

```python
df.info()
```

PREPARING AND CLEANING THE DATA

```python
df=df.drop(columns=["Unnamed: 0"])
```

```python
df
```

```python
df=df.dropna()
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.Sales.describe()
```

```python
x=df.iloc[:,:3]
y=df.iloc[:,-1]
```

SPLIT THE DATA INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
x_train.info()
```

```python
y_train.dtype
```

TRAIN THE DATA ON LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
l_reg=LinearRegression()
```

```python
l_reg.fit(x_train,y_train)
```

```python
l_reg_pred=l_reg.predict(x_test)
```

```python
l_reg_pred.reshape(-1,1)
```

ACCURACY OF DATASET

```python
from sklearn.metrics import r2_score
```

```python
r2_score(l_reg_pred,y_test)
```

VISUALISATION ON THE BASIS OF SALES 

```python
plt.scatter(df.iloc[:,0],df.iloc[:,-1],color="blue",label="TV")
plt.scatter(df.iloc[:,1],df.iloc[:,-1],color="red",label="Radio")
plt.scatter(df.iloc[:,2],df.iloc[:,-1],color="black",label="Newspaper")
plt.legend(loc="lower right")
plt.title("Sales vs. Advertising")
plt.xlabel("Advertising [1000 inr]")
plt.ylabel("Sales [Thousands of units]")
plt.grid()
plt.show()
```
