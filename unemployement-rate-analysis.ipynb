IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import plotly.express as px
```

READ THE DATA FROM PANDASREAD THE DATA FROM PANDAS

```python
df=pd.read_csv("/kaggle/input/unemployment-in-india/Unemployment_Rate_upto_11_2020.csv")
```

PREAPARING AND CLEANING THE DATA

```python
from sklearn import preprocessing
  
label_encoder = preprocessing.LabelEncoder()
  
df['Region']= label_encoder.fit_transform(df['Region'])  
df['Region'].unique()

df['Region.1']= label_encoder.fit_transform(df['Region.1'])  
df['Region'].unique()

df[' Frequency']= label_encoder.fit_transform(df[' Frequency'])  
df[' Frequency'].unique()

df[' Date']= label_encoder.fit_transform(df[' Date'])  
df[' Date'].unique()
df
```

```python
r=df.drop(" Estimated Unemployment Rate (%)",axis=1)

x=r.iloc[:,:]
y=df.iloc[:,3]
```

SPLIT THE DATA INTO TRAIN AND TEST 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

TRAIN THE MODEL ON LINEAR REGRESSION 

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

FIT THE MODEL

```python
LR.fit(x_train,y_train)
```

ACCURACY OF DATA

```python
reg=LR.predict(x_test)
LR.score(x_test,y_test)*1000
```
