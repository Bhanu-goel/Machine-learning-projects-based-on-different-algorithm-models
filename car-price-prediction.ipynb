CONVERT STRING DATA TYPE TO INT TYPE USING LABEL ENCODER

```python
from sklearn import preprocessing
  
label_encoder = preprocessing.LabelEncoder()
  
df['make_model']= label_encoder.fit_transform(df['make_model'])  
df['make_model'].unique()

df['body_type']= label_encoder.fit_transform(df['body_type'])  
df['body_type'].unique()

df['Body Color']= label_encoder.fit_transform(df['Body Color'])  
df['Body Color'].unique()

df['Gearing Type']= label_encoder.fit_transform(df['Gearing Type'])  
df['Gearing Type'].unique()

df['Extras']= label_encoder.fit_transform(df['Extras'])  
df['Extras'].unique()

df
```

```python
x=df.iloc[:,:7]
y=df.iloc[:,7]
```

SPLIT THE DATA INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40)
```

TRAIN THE DATA ON LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
#cross validation
from sklearn.model_selection import cross_val_score 
```

```python
regression = LinearRegression()
regression.fit(x_train,y_train)
```

ACCURACY OF DATA

```python
regression.score(x_test,y_test)
```

```python
regression.intercept_
```

```python
regression.coef_
```

```python
y_pred=regression.predict(x_test)
y_pred
```

PREDICTION OF CAR PRICE

```python
y_pred=regression.intercept_ + regression.coef_ * x
y_pred
```

TRY TO IMPROVE OUR DATASET USING RIDGE AND LASSO

```python
from sklearn.linear_model import Ridge , Lasso
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regression= Ridge()
ridge_regression
```

```python
parameters={"alpha": [1,2,10,20,30,40,50,60]}
ridgecv= GridSearchCV(ridge_regression,parameters, scoring= "neg_mean_squared_error", cv=5)
ridgecv.fit(x_train,y_train)
```

```python
ridge_pred =ridgecv.predict(x_test)
```

ACCURACY OF RIDGE MODEL

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,ridge_pred)
score
```

ACCURACY OF lASSO MODEL

```python

print("\nLasso Model............................................\n")
lasso = Lasso(alpha = 10)
lasso.fit(x_train,y_train)
train_score_ls =lasso.score(x_train,y_train)
test_score_ls =lasso.score(x_test,y_test)

print("The train score for ls model is {}".format(train_score_ls))
print("The test score for ls model is {}".format(test_score_ls))
```
